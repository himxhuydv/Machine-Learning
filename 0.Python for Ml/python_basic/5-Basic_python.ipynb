**FOR LOOP**:Iterate over the sequence of item through the iterable.  
1.Default start from 0.  
2.iterables last one is excluded mean if you gave the range 5 there will be the output for the number as 0,1,2,3,4,.  

```python
# Loop  
# FOR LOOP Syntax:
# Range of function 
for i in range(5): #i is temporary variable which store the iteration and print that 
    print(i)
```

**WHY DOES THE LAST VALUE GET's EXCLUDED?**
Because this is design in this term  
* start-included  
* end-excluded   
And if the last get excluded  that make the whole code mathmatically wrong if give wanna print 5 number it will start from the 0,1,2,3,4,5 it's now 6 digit not the 5 digit you wanna print

```python
for i in range(1,10,2):
    print(i)
#here as we have 2 as the step start from 1 then take two step ahead then do iteration and put the value into the temp variable.
```

```python
#Reverse a number for that we can easily the step in that -1 for backward as you know if you want to go for forward there  will be +1 and now for the backward that will be -1.
for i in range(10,1,-1): # we want is start the value from the 10 and goes in the reveverse upto 1 
    print(i)
```
