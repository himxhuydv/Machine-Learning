# Logical Operators

# AND: 
both the A and B needs to be True to get the output as True.if any of it having False it will consider as False in total.

# OR:
either one of the A and B if true whole statement is consider as True.

# NOT:

Not is equal to mean that if True then it will eventually mean as False and False will eventually be True.

```python
# Logical operators
# And,Or,Not
X=True
Y=False
C=True
```

```python
#And 
print(X and Y)
print(X and C)
```

```python
#Or
print(X or Y)
```

```python
#Not
print(X,not Y)  # so x is true and Y is false it  will enventually give true even it was false because uses the = right (not=!=)it contain the and = so true and false= false but in the not it will give the true.
#inverse way
print(not X) #if the x is true will be given as False
```
