# LIST:  
it's a data type which is mutable and stored in order.this is very usefull because we can acess the data just by the index.(which is a very key feature to know.)

```python
# List
freinds=[1,2,3,"himanshu","banana","avacado","Mango","papaya"]
# Acessing list we can directly acess the value just by the index.
print(freinds[2])
print(freinds[3])
#Negative indexing is also a thing that can very usefull in most cases.
print(freinds[-1]) # we can directly acess the last value.
```

```python
# Indexing
print(freinds[0:])
print(freinds[-1:-3]) # why this giving empty because the indexing works in the left to right but we if give the indeixng to right to left ot will print nothing
# overcoming this diffuclty print(freinds[-1:-3]) we have to use the step thing to over come that.
print(freinds[-1:-5:-1])# we have given step that is more important it will print the list in the reverse from left to rigtt
# over reveseral of the list 
print(freinds[::-1])
```

```python
# Modifying 
freinds[2]="watermalon" # we have updated the list at the index at 2 as the watermalon

print(freinds[2])
```

```python
# Modifying evreything using the indexes
freinds[1:]="Watermalon"
print(freinds) # everything is modifed and charater in the watermalon as the single digit which very impressive "w"."a","t" 

```

```python
# Append
freinds.append("data")
print(freinds)
```

```python
# wanna add element at specific index 
freinds.insert(2,"papaya")
print(freinds) 
# each time we run the code that will get's add on the list so now we have a way to ccrash this by removing these erorr 
```

```python
 # remove  function
freinds.remove("give the character you want to ennter") # it's remove the first occurance of the element
print(freinds)
```

```python
# Removing the last index and returing it 
poped_fruit=freinds.pop() #auto matically remove the last element from the list without givng any arguments which is more crazy 
print(poped_fruit)
```

```python
# Indexing  for knowing which index the object we are talking about comes
freinds.index("papaya")

```

```python
#count how much any element is occur that a very important thing to know
freinds.insert(2,"papaya")

freinds.count("papaya") # it's just counting how many time a thing's comes

freinds[0]="chivi"
freinds[1]="leechi"
print(freinds[9])
freinds.remove("papaya")
print(freinds)
```

```python
# sort function which basically sort the list in the acesending order if there is only the character it will gonna do based on the character that's comes
print(freinds)
freinds.sort()
print(freinds)
# now we do the sort
```

```python
 # Reverse function this automatically reverse the whole list
freinds.reverse()
print(freinds)
```

```python
# slicing 
print(freinds[::])
print(freinds[1:5])
```

```python
# iteration how can we do
for i in freinds:
    print(i)
```

```python
# iteration can we print index 
for index,i in enumerate(freinds):
    print(index,i)
```
