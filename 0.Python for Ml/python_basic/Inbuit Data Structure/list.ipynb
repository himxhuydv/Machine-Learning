# LIST:  
it's a data type which is mutable and stored in order.this is very usefull because we can acess the data just by the index.(which is a very key feature to know.)

```python
# List
freinds=[1,2,3,"himanshu","banana","avacado","Mango","papaya"]
# Acessing list we can directly acess the value just by the index.
print(freinds[2])
print(freinds[3])
#Negative indexing is also a thing that can very usefull in most cases.
print(freinds[-1]) # we can directly acess the last value.
```

```python
# Indexing
print(freinds[0:])
print(freinds[-1:-3]) # why this giving empty because the indexing works in the left to right but we if give the indeixng to right to left ot will print nothing
# overcoming this diffuclty print(freinds[-1:-3]) we have to use the step thing to over come that.
print(freinds[-1:-5:-1])# we have given step that is more important it will print the list in the reverse from left to rigtt
# over reveseral of the list 
print(freinds[::-1])
```

```python
# Modifying 
freinds[2]="watermalon" # we have updated the list at the index at 2 as the watermalon

print(freinds[2])
```

```python
# Modifying evreything using the indexes
freinds[1:]="Watermalon"
print(freinds) # everything is modifed and charater in the watermalon as the single digit which very impressive "w"."a","t" 

```

```python
# Append
freinds.append("data")
print(freinds)
```

```python
# sort function which basically sort the list in the acesending order if there is only the character it will gonna do based on the character that's comes
print(freinds)
freinds.sort()
print(freinds)
# now we do the sort
```

### List Comprehension  
we have two important thing:  
* Mapping   
* Filtering    
1.Basic syntax   **[expression(any kind of expression could be that),for item in iterable**]        
2.with conditional Logic   **[expression for item in iterable If condition is required**]   
3.Nested list Comprehension   **[expression iterating with item 1 and iterating with item 2]**

**Mapping**:what action will perform or opertion you want to do for example x**2 which is very important thing to know.  
**Filtering**:here we do filtering thing we do the iteration thing


```python
number=[i for i in  range(0,11)]
print(number)
```

```python
# List 
lis_1=[x**2 for x in range(10)]
print(lis_1)
```

### List Comprehension
1.Basic Syntax [expression(x**2)  for item in iterable    If condition is required  ]  
2.Nested list Comprehension  
3.List Comprehension with function   
  And you can addup this thing if you want to do more complex things

```python
# list comprehension

# Basic list comprehension 
number=[x**2 for x in range(0,10)]
print(number)
```

**Trying to Find number between a range  only the even number.**

Normal way's

```python
# first we try with the normal way's
lis=[]
for i in range(1,10):
    lis.append(i)
print(lis)
```

#### List comprehension way 

```python
even=[x for x in range(0,10)if x%2==0]
print("even number",even)
```

Nested List Comprehension:iterate two more times


```python
# Nested list comprehensionm
number1=[(x,i) for x in range(1,11) for i in range(20,31)]
print(number1)
```
