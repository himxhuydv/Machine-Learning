Question 9: Write a program that print all the number from 1 to 10 except 5 using a for loop and continue statement.

```python
# Answer
for i in range(1,11):
    if i==5:
        continue
    else:
        print(i)
```

Question 12: Write a program that calculates the factorial of a number input by the user using a while loop.

```python
input_user=int(input("Enter the number you want to take factorial "))
fact=1
for i in range(1,input_user+1):
    fact=i*fact
    print(fact)
```

Question 13: Write a program that calculates the sum of the digits of a number input by the user using a while loop.

```python
user_input=int(input("Enter the number"))
count=0
sum=0
while count<user_input+1:
    sum+=count
    count+=1
print(sum)
```
