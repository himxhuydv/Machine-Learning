## Class  
class basically mean the blueprint or  structure for creating the object.    
It's generally mean the class would have feature from the object (**In the simple term we can say's that the class contain the atributes and method of the object.**)


````python
#class 
class Student:
    name="karan"
#creating object(instances)
s1=Student() #we are creating the object using the instance(class)
print(s1.name)# we are acessing the the class attributes
s2=Student()
print(s2.name)
````

**Class**
Now we are gonna make a factory in that we are gonna produce car


````python
#Class of car (what actually that car need's to have.)
class Car:
    colour="black"
    brand="bmw"
    wheels=4
BMW=Car()
print(BMW.colour)
print(BMW.brand)
print(BMW.wheels)
````
