**CONDITIONAL STATEMENT:** this usually mean that there has been some condition.if that thing is true then execute this block of code.   
There are type of Conditional Statement:
* IF STATEMENT
* ELSE STATEMENT
* ELIF STATEMENT

**IF STATEMENT:**

IF STATEMENT:  Usually stand for this.ifthis is true then this will happened.

```python
age=20
if age>=18:
    print("You are male:")
```

**ELSE STATEMENT:**

ELSE STATEMENT: It's basically mean when if statement is not rigth then it.this will happened instead of if statement we execute else statement.

```python
age=9
if age>=18: #this statement become false (because the age is not less than 18.)
    print("You are adult")
else:
    print("you teenager ") #this condition become true. 
```

ELIF STATEMENT:  **IF STATEMENT IS TRUE THEN TRY THIS STATEMENT TO.**
  
Means there has been multiple statment right we have multiple statement avaiable that where we use the *elif statement.*

**ELIF STATEMENT:**

```python
age=19
if age>=17:
    print("You have sucessfully passed the legal marriage.")
elif age>=13:
    print("You are big man.")
else:
    print("You are not big man.")
```

**CONDITIONAL STATEMENT:**


CONDITIONAL STATEMENT:Usually mean that we have mulitple if and elif statement.If statement inside that if statement there is another if statement,and can be inside that if an if statement,it's can an another if statement and so on.

```python
#POSITIVE,ODD,EVEN,NEGATIVE:
number=int(input("Enter the number you want to check:"))
if number>0:
    print("Postive")
    if number%2==0:
        print("Even")
    else:
        print("Odd")
elif (number==0):
    print("zero")
else:
    print("Negative")
```
