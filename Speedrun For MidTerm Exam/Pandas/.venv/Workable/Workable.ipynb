```python
import pandas as pd 
dict1={"Team":["Barca","Fcb","arsenal","manchester united"],"Score":[1,2,4,5]}
df=pd.DataFrame(dict1,index=["20/07/2025","21/08/2025","22/09/2025","01/02/2025"])
print(df)
#df.to_csv("football record.csv")#after this code this will automatically saved the file into the workable section
# Hide the unwanted index using the index=False
df.to_csv("football team.csv",index=False)
# for the jason file
#df.to_json("football record.json")
```
